```python
import streamlit as st
import google.generativeai as genai

genai.configure(api_key="ENTER_YOUR_API_KEY")
model = genai.GenerativeModel("gemini-2.5-flash")
st.title("Jarvis Chatbot")

if "chat" not in st.session_state:
    st.session_state.chat = []

for role, msg in st.session_state.chat:
    if role == "ME":
        st.chat_message("ME").markdown(msg)
    else:
        st.chat_message("Jarvis").markdown(msg)

user = st.chat_input("Ask me anything...")

if user:
    st.session_state.chat.append(("ME", user))
    st.chat_message("ME").markdown(user)
    response = model.generate_content(user)
    st.session_state.chat.append(("Jarvis", response.text))
    st.chat_message("Jarvis").markdown(response.text)
```
